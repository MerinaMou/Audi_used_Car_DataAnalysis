## Exploratory Data Analysis (EDA) using groupby/pivot_table

#### Importing libraries and packages

```python
import numpy as np 
import pandas as pd 

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context("talk")
sns.set_style("darkgrid")
```

```python
df = pd.read_csv(r'C:\Users\USER\Downloads\audi.csv')
#Display the Data Set
display(df.head(5))
display(df.tail(5))
```

```python
#Data statistics
display(df.describe())

# statistics of the object variables
display(df.describe(include='object'))

#checking for data type
print (df.info())
```

### EDA based on car Model

```python
#check the number of unique values in the 'model' column
print(df['model'].unique())

#Groupby between 'model' and 'price' column
df_agg_model = df.groupby('model')['price'].agg(['count', 'mean', 'sum'])
display(df_agg_model.head())
```

### Data processing before visualization by Model

```python
# sort the dataframe based on the column 'count'
df_agg_model = df_agg_model.sort_values(by='count', ascending=False)

# create a new column 'percentage of cars' of each model
df_agg_model['percent_of_cars'] = (df_agg_model['count']/df_agg_model['count'].sum())*100

# convert the 'sum' column in millions
df_agg_model['sum'] = df_agg_model['sum']/1000000

# keep up to two decimal points
df_agg_model = df_agg_model.round(2)

# remove the 'count' column
df_agg_model = df_agg_model.drop('count', axis=1)

display(df_agg_model.head())
```

### Barplot: Percentage of cars, Average and Total selling price of cars by model

```python
plt.figure(figsize=(20,8))

plt.subplot(1, 3, 1)
sns.barplot(x='percent_of_cars', y=df_agg_model.index, data=df_agg_model, orient='h', color='yellow')
plt.title("Percentage of cars in each model")
plt.xlabel("Percentage of cars (%)")
plt.ylabel("model")


plt.subplot(1, 3, 2)
sns.barplot(x='mean', y=df_agg_model.index, data=df_agg_model, orient='h', color='green')
plt.title("Average selling price of cars by each model")
plt.xlabel("Average selling price ($)")
plt.ylabel("")


plt.subplot(1, 3, 3)
sns.barplot(x='sum', y=df_agg_model.index, data=df_agg_model, orient='h', color='purple')
plt.title("Total selling price of cars by each model")
plt.xlabel("Total selling price ($ million)")
plt.ylabel("")


plt.suptitle("Barplot of the AUDI used cars show Percentage of cars, Average and Total selling price of cars by model")
plt.tight_layout()
plt.show()
```

#### Findings:
1. Top 5 selling car models are **A3 (18.08%), Q3 (13.28%), A4 (12.95%), A1 (12.63%) and A5 (8.27%)**
2. Top 5 average selling price car models are **R8 ($97,652), Q8 ($51,265), and RS4 ($50, 152)**
3. Top 5 total selling price (Million doller) car models are **A3 (33.58M), Q3 (32,59M), A4 (27,97M), Q5 (26.07M), and A5 (20.80M)**

### EDA based on Transmission

```python
#check the number of unique values in the 'transmission' column

print(df['transmission'].unique())

#Groupby between 'transmission' and 'price' column
df_agg_transmission = df.groupby('transmission')['price'].agg(['count', 'mean', 'sum'])
display(df_agg_transmission)
```

### Data processing before visualization by Transmission

```python
# sort the dataframe based on the column 'count'
df_agg_transmission = df_agg_transmission.sort_values(by='count', ascending=False)

# create a new column 'percentage of cars' in each model
df_agg_transmission['percent_of_cars'] = (df_agg_transmission['count']/df_agg_transmission['count'].sum())*100

# convert the 'sum' column in millions
df_agg_transmission['sum'] = df_agg_transmission['sum']/1000000

# keep up to two decimal points
df_agg_transmission = df_agg_transmission.round(2)

# remove the 'count' column
df_agg_transmission = df_agg_transmission.drop('count', axis=1)

display(df_agg_transmission)
```

### Barplot: Percentage of cars, Average and Total selling price of cars by transmission

```python
plt.figure(figsize=(20,8))

plt.subplot(1, 3, 1)
sns.barplot(x='percent_of_cars', y=df_agg_transmission.index, data=df_agg_transmission, orient='h', color='Violet')
plt.title("Percentage of cars in each transmission")
plt.xlabel("Percentage of cars (%)")
plt.grid()

plt.subplot(1, 3, 2)
sns.barplot(x='mean', y=df_agg_transmission.index, data=df_agg_transmission, orient='h', color='green')
plt.title("Average selling price of cars by each transmission")
plt.xlabel("Average selling price ($)")
plt.ylabel("")

plt.subplot(1, 3, 3)
sns.barplot(x='sum', y=df_agg_transmission.index, data=df_agg_transmission, orient='h', color='blue')
plt.title("Total selling price of each transmission")
plt.xlabel("Total selling price ($ millions)")
plt.ylabel("")

plt.suptitle("Barplot of the AUDI used cars show Percentage of cars, Average and Total selling price of cars by transmission")
plt.tight_layout()
plt.show()
```

#### Findings:
1. There are 3 different types of transmission type: **Manual, Semi-auto, and Automatic.**
2. The barplots illustrate that the dataset contains most amount of data on **Manual cars (40.95%**.
3. The average selling price of a car by transmission type shows that **Automatic cars** have the **largest selling price**.
4. Total selling price (million dollers) of **Semi-auto cars 97.54M, Automatic 76.38M, and Manual 70.35M.**
